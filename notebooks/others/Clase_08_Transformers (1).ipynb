# Sentiment Analysis

```python
sa = pipeline('sentiment-analysis')
```

```python
sa("We are happy to be in last session of the course")
```

```python
sa("The movie was really bad")
```

```python
sa("I don't mind walking")
```

```python
sa_es = pipeline('sentiment-analysis', model="nlptown/bert-base-multilingual-uncased-sentiment")
```

```python
sa_es("El producto es excelente")
```

```python
sa_es("No me gustó mucho la comida")
```

# Question Answering

```python
qa = pipeline('question-answering')
```

"The bicentennial of the Independence of Peru will occur on 28 July 2021. Its celebration will commemorate 200 years since Peru's proclamation of Independence. The celebration will consist of a mix of local, state, and national activities that have been being planned since 2016. On 8 August 2016, the Prime Minister of Peru announced the creation of the Organizing Commission for the Commemoration of the Bicentennial of the Independence of Peru.[1]"

```python
context = "The bicentennial of the Independence of Peru will occur on 28 July 2021. Its celebration will commemorate 200 years since Peru's proclamation of Independence. The celebration will consist of a mix of local, state, and national activities that have been being planned since 2016. On 8 August 2016, the Prime Minister of Peru announced the creation of the Organizing Commission for the Commemoration of the Bicentennial of the Independence of Peru.[1]"
```

```python
qa(question="What will occur on 28 July 2021?", context=context)
```

```python
qa(question="What will the celebration conmemorate?", context=context)
```

```python
qa(question="Who is the prime minister?", context=context)
```

# Named Entity Recognition

```python
ner = pipeline("ner")
```

"""Bitcoin has jumped past $30,000 as Elon Musk said Tesla is "most likely" to start accepting it as payment again."""

```python
text = """Bitcoin has jumped past $30,000 as Elon Musk said Tesla is "most likely"
          to start accepting it as payment again."""
```

```python
[print(r) for r in ner(text)]
```
